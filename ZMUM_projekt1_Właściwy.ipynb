### Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder

def encode_ordinal(train_set,chosen_columns):
    
    ord_enc = OrdinalEncoder()
    
    encoded = ord_enc.fit_transform(train_set[chosen_columns])
    
    train_set.loc[:,chosen_columns] = encoded
    
    return train_set
```

```python
train = encode_ordinal(train, all_categorical_columns)
train.head(10)
```

## Przygotowanie miary prec@10

```python
def prec_10(y, y_preds):
    
    pred_results = pd.concat([pd.Series(list(y.index)),pd.Series(y_preds[:,1])],axis=1).set_index(0)
    pred_results = pred_results.rename(index=int, columns={1: "proba_value"})
    pred_results.index.names = ['index']
    pred_results = pred_results.sort_values(by=['proba_value'],ascending=False)
    
    n = int(len(pred_results)*0.1)
    pred_10_results = pred_results.iloc[0:n]  
    
    pred_10_results['class'] = 1
    pred_10_results = pred_10_results.drop(columns='proba_value')
    
    y_df = pd.DataFrame({'index':y.index, 'class':y.values}).set_index('index')
    
    yhat = len(pred_10_results)
    
    matches = 0

    for element in list(pred_10_results.index): 
        if y_df.loc[element]['class']==1:
            matches += 1
    
    #print(matches)
    #print(yhat)
    
    acc_value = matches/yhat
    
    return acc_value
```

## KaÅ¼dy z modeli jest obudowany funkcjÄ… random search, ktÃ³ra pozwala wybraÄ‡ parametry dajÄ…ce najwyÅ¼szy wynik wg miary prec@10

## Model 1: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
import random

final_prec = 0
depths = list(range(1, 30))
max_features_type = ['auto','log2',None]
n_est = list(range(1, 10))

for i in range(300):
    n = random.choice(n_est)
    m_d = random.choice(depths)
    m_f = random.choice(max_features_type)
    
    random_forest = RandomForestClassifier(n_estimators=n*50, max_depth=m_d,max_features=m_f,random_state=0)
    y_pred_proba = random_forest.fit(X_train,y_train).predict_proba(X_test)
    prec10 = prec_10(y_test, y_pred_proba)

    if(final_prec<prec10):
        final_prec = prec10
        chosen_n_estim = n*50
        chosen_max_depth = m_d
        chosen_max_features = m_f

        print("Prec@10: " + str(prec10))
        print("chosen_n_estim: " + str(chosen_n_estim))
        print("chosen_max_depth: " + str(chosen_max_depth))
        print("chosen_max_features: " + str(chosen_max_features))
        print("------------")
    
    if(i%25==0):
        print(i)
    
    del(random_forest)
```

## Model 2: XGBoost

```python
from xgboost import XGBClassifier
import random

final_prec = 0
n_est = list(range(1, 10))
depths = list(range(1, 30))
l_rates = [0.5]+list(range(1, 5))
gammas = list(range(0,5))

for i in range(450):
    n = random.choice(n_est)
    m_d = random.choice(depths)
    lr = random.choice(l_rates)
    g = random.choice(gammas)
    
    xgb = XGBClassifier(n_estimators=n*50, max_depth=m_d,learning_rate=lr*0.1,gamma=g,random_state=25)
    y_pred_proba = xgb.fit(X_train,y_train).predict_proba(X_test)
    prec10 = prec_10(y_test, y_pred_proba)

    if(final_prec<prec10):
        final_prec = prec10
        chosen_n_estim = n*50
        chosen_max_depth = m_d
        chosen_lr = lr*0.1
        chosen_gamma = g

        print("Prec@10: " + str(prec10))
        print("chosen_n_estim: " + str(chosen_n_estim))
        print("chosen_max_depth: " + str(chosen_max_depth))
        print("chosen_lr: " + str(chosen_lr))
        print("chosen_gamma: " + str(chosen_gamma))
        print("------------")
    
    if i%25 == 0:
        print(i)
    
    del(xgb)
```

## Model 3: AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import random

final_prec = 0
n_est = list(range(1, 10))
depths = list(range(1, 20))
alghorithms = ['SAMME', 'SAMME.R']
l_rates = [0.5]+list(range(1, 5))


for i in range(100):
    m_d = random.choice(depths)
    lr = random.choice(l_rates)
    alg = random.choice(alghorithms)
    n = random.choice(n_est)
    
    adab = AdaBoostClassifier(DecisionTreeClassifier(max_depth=m_d),
                         learning_rate=lr*0.1,
                         algorithm=alg,
                         n_estimators=n*50)
    y_pred_proba = adab.fit(X_train,y_train).predict_proba(X_test)
    prec10 = prec_10(y_test, y_pred_proba)

    if(final_prec<prec10):
        final_prec = prec10
        chosen_n_estim = n*50
        chosen_max_depth = m_d
        chosen_algthoritm = alg
        chosen_lr = lr*0.1

        print("Prec@10: " + str(prec10))
        print("chosen_n_estim: " + str(chosen_n_estim))
        print("chosen_max_depth: " + str(chosen_max_depth))
        print("chosen_algthoritm: " + str(chosen_algthoritm))
        print("chosen_lr: " + str(chosen_lr))
        print("------------")
    
    if(i%25==0):
        print(i)
    
    del(adab)
```
